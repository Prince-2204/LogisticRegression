# Data Cleaning

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["RowNumber" , "CustomerId" , "Surname"] , inplace = True)
```

```python
df.head()
```

```python
def outlier(df , column):
    sorts = df[column].sort_values()
    q1 = sorts.quantile(0.25)
    q3 = sorts.quantile(0.75)
    iqr = q3 - q1
    lower_bound = sorts < (q1 - 1.5*iqr)
    upper_bound = sorts > (q3+1.5*iqr)
    filtered_data = df[~((lower_bound)|(upper_bound))].reset_index(drop=True)
    
    return filtered_data
```

```python
df = outlier(df,"CreditScore")
```

```python
df = outlier(df , "Age")
```

```python
df = outlier(df , "Tenure")
```

```python
df = outlier(df , "Balance")
```

```python
df = outlier(df ,"NumOfProducts" )
```

```python
df = outlier(df , "HasCrCard")
```

```python
df = outlier(df , "IsActiveMember")
```

```python
df = outlier(df , "EstimatedSalary")
```

```python
df.info()
```

```python
from sklearn import preprocessing
```

```python
for col in df.select_dtypes(include=['object']).columns:
    labelencoder = preprocessing.LabelEncoder()
    labelencoder.fit(df[col].unique())
    df[col] = labelencoder.transform(df[col])
    
```

```python
df
```

```python
df.info()
```

```python
X = df.drop("Exited" , axis=1)
y = df["Exited"]
```

```python
def featurescaling(X):
    Xmean = X.mean(axis=0)
    Xstd = X.std(axis=0)
    
    return (X-Xmean)/Xstd
```

```python
X = featurescaling(X)
```

```python
X.mean(axis=0)
```

```python
X.std(axis=0)
```

# Feature Selection

```python
from sklearn.feature_selection import SelectKBest
```

```python
best_features = SelectKBest()
fit = best_features.fit(X,y)
```

```python
fit.scores_
```

```python
dffeatures = pd.DataFrame(X.columns)
dfscores = pd.DataFrame(fit.scores_)
featurescores = pd.concat([dffeatures , dfscores] , axis=1)
featurescores.columns = ["Feature Name" , "Scores"]

featurescores.sort_values(by="Scores" , ascending = False)
```

```python
X_new = df[["Age" , "IsActiveMember" , "Balance" , "NumOfProducts" , "Gender"]]
```
