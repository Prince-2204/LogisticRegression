```python
# Removing The outliers
def outlier(df,columns):
    sorts = df[columns].sort_values()
    q1 = sorts.quantile(0.25)
    q3 = sorts.quantile(0.75)
    iqr = q3-q1
    lower_bound = sorts < (q1-1.5*iqr)
    upper_bound = sorts>(q3 + 1.5*iqr )
    filtered_data = df[~((lower_bound)|(upper_bound))].reset_index(drop=True)
    
    return filtered_data
```

```python
df = outlier(df , 'BMI')
```

```python
df = outlier(df , 'PhysicalHealth')
```

```python
df = outlier(df , 'MentalHealth')
```

```python
df = outlier(df , 'SleepTime')
```

```python
y = df['HeartDisease']                # Extracting all the labels
```

# Feature Scaling

```python
# Defining the the function to do the featurescaling to make the values in similar range 
def featurescaling(X):
    Xmean = X.mean(axis=0)
    Xstd = X.std(axis=0)
    
    return (X - Xmean)/Xstd
```

```python
X = featurescaling(X)
```

```python
X.mean(axis=0)
```

```python
X.std(axis=0)
```

# Feature Selection

```python
from sklearn.feature_selection import SelectKBest
```

```python
# Selecting Best Features using sklearn
best_features = SelectKBest()
Fit = best_features.fit(X,y)
```

```python
Fit.scores_
```

```python
dfscores = pd.DataFrame(Fit.scores_)
dfcoluns = pd.DataFrame(X.columns)

featurescores = pd.concat([dfcoluns,dfscores],axis = 1)
featurescores.columns = ['Feature Name' , 'Scores']
```

```python
featurescores
```

```python
featurescores.sort_values(by="Scores" , ascending = False)
```

```python
X_new = df[['AgeCategory' , 'Stroke' ]]  # Selecting features on the basis of Feature scores
```
